Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

```python
import pandas as pd

confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

confirmed_cases = pd.read_csv(confirmed_cases_url)
deaths = pd.read_csv(deaths_url)

global_confirmed_cases = confirmed_cases.iloc[:, 4:].sum(axis=0)
global_deaths = deaths.iloc[:, 4:].sum(axis=0)

global_data = pd.DataFrame({'Confirmed Cases': global_confirmed_cases, 'Deaths': global_deaths})

global_data
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['SPY', 'XIU.TO', 'DAL', 'SPG', 'GOLD']

confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
confirmed_cases = pd.read_csv(confirmed_cases_url)
deaths = pd.read_csv(deaths_url)

global_cases = confirmed_cases.iloc[:, [0, 1, -1]].groupby('Country/Region').sum()
global_deaths = deaths.iloc[:, [0, 1, -1]].groupby('Country/Region').sum()
global_data = global_cases.join(global_deaths, lsuffix='_cases', rsuffix='_deaths')
global_data['Total_cases'] = global_data.iloc[:, -2].astype(int)
global_data['Total_deaths'] = global_data.iloc[:, -1].astype(int)

stock_data = {}

for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist_data = stock.history(period="1d")
    daily_high = hist_data['High'][0]
    daily_low = hist_data['Low'][0]
    stock_data[ticker] = {'Daily High': daily_high, 'Daily Low': daily_low}

for ticker in tickers:
    global_data[ticker + '_Daily_High'] = stock_data[ticker]['Daily High']
    global_data[ticker + '_Daily_Low'] = stock_data[ticker]['Daily Low']
    
global_data
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['SPY', 'XIU.TO', 'DAL', 'SPG', 'GOLD']

for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist_data = stock.history(start='2019-01-01', end='2023-12-31')
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.plot(hist_data['Close'])
    ax.set(title=ticker + ' 2019 to 2023 Data', xlabel='Date', ylabel='Closing Price')
    plt.show()

```

*   It enables us to see the market's overall trend and whether there have been any notable changes in the numbers, we could also plot the daily high and low values for each stock market index and for gold over time.In above graph we can see the effects on stock prices during the COVID Period and notice the trends and patterns.


*   In conclusion,a line graph is a helpful tool for visualising this kind of data since it may show the patterns and trends that emerge over time, making it simpler to see any significant alterations or advancements.
